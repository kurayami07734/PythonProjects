```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
#Load the data
data = pd.read_csv('./real_estate_price_size_year.csv')
```

```python
data.describe()
```

```python
x1 = data[['size', 'year']]
y = data['price']

x = sm.add_constant(x1)
res = sm.OLS(y, x).fit()
res.summary()
```

```python
#3rd exercise with dummy variables
raw_data = pd.read_csv('./real_estate_price_size_year_view.csv')
raw_data.head()
```

```python
data = raw_data.copy()
data['view'] = data["view"].map({"Sea view": 1, "No sea view": 0})
data.describe()
```

```python
x1 = data[['size', 'year', 'view']]
y = data['price']

x = sm.add_constant(x1)
res = sm.OLS(y, x).fit()
res.summary()
```

```python
new_data = pd.DataFrame({"const": 1, "size": [104000.98], "year": [2023], "view" : [1]})
predictions = res.predict(new_data)
predictions
```

```python
predDf = pd.DataFrame({"predicted price": predictions})
joined = new_data.join(predDf)
joined
```
