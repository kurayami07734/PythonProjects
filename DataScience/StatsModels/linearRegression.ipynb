## Importing libs

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

## Loading data

```python
data = pd.read_csv("./simpleLinearRegression.csv")
data.describe()
```

```python
y = data['GPA']
x = data['SAT']

plt.scatter(y, x)
plt.xlabel("SAT", fontsize=20)
plt.ylabel("GPA", fontsize=20)
plt.show()
```

```python
# regression
x1 = sm.add_constant(x)
results = sm.OLS(y, x1).fit()
results.summary()
```

```python
plt.scatter(x, y)
yhat = 0.0017 * x + 0.275
fig = plt.plot(x, yhat, lw=4, c='orange', label='regression line')
plt.xlabel("SAT", fontsize=20)
plt.ylabel("GPA", fontsize=20)
plt.show()
```

```python
## Real Estate price Exercise

data = pd.read_csv('./real_estate_price_size.csv')
y = data['price']
x = data['size']

plt.scatter(y, x)
plt.xlabel('Size', fontsize=20)
plt.ylabel('Price', fontsize=20)
plt.show()
```

```python
x1 = sm.add_constant(x)
res = sm.OLS(y, x1).fit()
res.summary()
```

```python
yhat = 223.1787 * x + 1.019e5
plt.scatter(y, x)
fig = plt.plot(yhat, x, lw=2, c='red', label='regression label')
plt.xlabel("Size", fontsize=20)
plt.ylabel("Price", fontsize=20)
plt.show()
```
